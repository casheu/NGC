### Hyphothesis Testing

Apabila kita ingin mencari apakah ada penurunan yang signifikan dalam kasus baru sejak bulan Juli tahun ini misalnya di Bali, kita dapat menggunakan hypothesis testing menggunakan single sample test dengan hypothesis:<br>
H0 : μ1th >= μ3th  (tidak ada penurunan)<br>
H1 : μ1th < μ3th (ada penurunan)

Pertama, kita pisahkan data Bali bulan Juli tahun ini hingga sekarang.

```python
bali = newdf[newdf['Location'] == 'Bali']
newbali = bali[bali['Date'] >= '2022-07-01']
newbali
```

Kemudian kita coba hitung rata-rata harian masing-masing data tersebut.

```python
# Menghitung beban harian rata-rata
daily_A = float(bali[['Date','New Cases']].mean())
daily_B = float(newbali[['Date','New Cases']].mean())
daily_A, daily_B
```

Dengan perhitungan mean terlihat ada penurunan, namun untuk mengetahui apakah penurunan ini signifikan, kita dapat melanjutkan mencari p-value.

```python
t_stat,p_val = stats.ttest_1samp(newbali['New Cases'], daily_A)
print('P-value:',p_val/2) #The p-value divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat)
```

Terlihat p-value sangat kecil dan dipastikan dibawah 0.05 maka kita dapat mereject h0, namun untuk memudahkan, kita lakukan visualisasi terlebih dahulu.

```python
daily_A_pop = np.random.normal(newbali['New Cases'].mean(), newbali['New Cases'].std(), 10000)

ci = stats.norm.interval(0.90, newbali['New Cases'].mean(), newbali['New Cases'].std())

plt.figure(figsize=(16,5))
sns.distplot(daily_A_pop, label='Daily Income (Population)', color='blue')
plt.axvline(newbali['New Cases'].mean(), color='red', linewidth=2, label='Daily Income (Mean)')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%')
plt.axvline(daily_A_pop.mean() + t_stat*daily_A_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
plt.legend()
```

Terlihat jelas bahwa garis h1 (garis hitam) berada diluar confidence threshold kita, sehingga dapta dikatakan h0 berhasil di reject, sehingga dapat disimpulkan bahwa ada penurunan jumlah kasus baru secara signifikan sejak bulan Juli tahun ini.
