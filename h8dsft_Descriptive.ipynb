### Central Tendency

Permasalahan pertama ialah memberi informasi berdasarkan central tendency (mean, median, dan modus) untuk masing-masing site.

Kita dapat mencari rata-rata beban yang diolah oleh masing-masing site.

```python
a = newdf.groupby('Location')['Total Cases'].mean()
a = a.sort_values(ascending=False)
a
```

Untuk mempermudah informasi, kita gunakan visualisasi dengan bar chart.

```python
plt.bar(a.index,a)
plt.title('Mean Masing-masing Site', fontsize=18)
plt.xlabel('Site', fontsize=10)
plt.ylabel('Weight', fontsize=10)
plt.xticks(rotation='vertical')
fig = plt.figure(figsize = (20, 10))
```

Dari chart ini dapat terlihat jelas mean dari masing-masing site, dari yang terbesar GREAT NORTHERN hingga yang terkecil AUSTIN IRON AND METAL

Selain mean ada juga median yang merepresentasikan value paling tengah dari setiap site.

```python
a = newdf.groupby('Location')['Total Cases'].median()
a = a.sort_values(ascending=False)
a
```

```python
plt.bar(a.index,a)
plt.title('Median Masing-masing Site', fontsize=18)
plt.xlabel('Site', fontsize=10)
plt.ylabel('Weight', fontsize=10)
plt.xticks(rotation='vertical')
fig = plt.figure(figsize = (20, 10))
```

Dan central tendency yang terakhir ialah mode, yakni value yang paling sering keluar dalam suatu distribusi.

```python
a = a = newdf.groupby('Location')['Total Cases'].apply(pd.Series.mode)
a
```

Dari sini terlihat bahwa untuk mode dari weight tiap site tidak memberikan informasi yang relevan, namun kita dapat menggunakan mode secara relevan dengan mencari load_type yang paling sering ditemukan dalam setiap site.

```python
a = newdf['Location']
a.mode()
```

Berdasarkan analisa central tendency yang kita lakukan sebelumnya, kita dapat memberi informasi berupa:
- Nilai rata-rata beban yang diolah oleh setiap site.
- Visualisasi perbandingan nilai rata-rata tersebut antar site.
- Nilai paling tengah beban yang diolah setiap site.
- Visualisasi perbandingan nilai tersebut antar site.
- Type beban yang paling sering ditemukan di masing-masing site.

### Variability

Kita akan lanjutkan analisa data TDS - MRF yang sudah tidak memiliki outlier dengan mencari range, variance, dan standar deviasi nya.

```python
total = newdf['Total Cases']

print('range =', str(total.max() - total.min()))
print('variance =', total.var())
print('standar deviasi =', total.std())
```

Informasi yang kita dapatkan ialah:
- Rentang jarak data weight yang kita dapatkan ialah 21860 ton, berdasarkan selisih dari weight terkecil dan weight terbesar.
- Jarak rata-rata antara beban dengan rata-rata beban keseluruhan ialah 3965,235 ton.
